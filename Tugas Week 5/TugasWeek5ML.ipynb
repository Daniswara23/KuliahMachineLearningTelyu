```python
# Import pustaka yang diperlukan
import pandas as pd              # Untuk manipulasi data
import numpy as np               # Untuk operasi numerik
import matplotlib.pyplot as plt   # Untuk visualisasi data
import seaborn as sns            # Untuk visualisasi yang lebih estetis

# Membaca dataset dari file CSV
data = pd.read_csv('/content/sample_data/data.csv')   # Mengimpor data dari file CSV ke dalam DataFrame

# Menampilkan 5 baris pertama dari dataset
print(data.head())                # Melihat data awal untuk memahami struktur dataset

# Mengetahui informasi dasar tentang dataset
print(data.info())                # Menampilkan informasi tentang tipe data dan jumlah non-null

# Menampilkan deskripsi statistik dari dataset
print(data.describe())            # Menghasilkan ringkasan statistik untuk kolom numerik

# Memeriksa missing values dalam dataset
print(data.isnull().sum())        # Menampilkan jumlah missing values untuk setiap kolom

# Visualisasi distribusi kolom tertentu
plt.figure(figsize=(10, 6))      # Mengatur ukuran plot
sns.histplot(data['kolom_anda'], kde=True)  # Membuat histogram dengan estimasi kepadatan
plt.title('Distribusi Kolom Anda') # Menambahkan judul
plt.xlabel('Nilai Kolom Anda')    # Menambahkan label sumbu x
plt.ylabel('Frekuensi')            # Menambahkan label sumbu y
plt.show()                        # Menampilkan plot

# Memeriksa hubungan antar kolom
plt.figure(figsize=(12, 8))       # Mengatur ukuran plot
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')  # Membuat heatmap dari matriks korelasi
plt.title('Matriks Korelasi')     # Menambahkan judul
plt.show()                        # Menampilkan plot

# Memvisualisasikan boxplot untuk mendeteksi outlier
plt.figure(figsize=(10, 6))       # Mengatur ukuran plot
sns.boxplot(x='kolom_kategorikal', y='kolom_numerik', data=data)  # Membuat boxplot
plt.title('Boxplot Kolom Kategorikal vs Numerik') # Menambahkan judul
plt.xlabel('Kolom Kategorikal')   # Menambahkan label sumbu x
plt.ylabel('Kolom Numerik')        # Menambahkan label sumbu y
plt.show()                        # Menampilkan plot

# Menyimpan DataFrame yang sudah diproses
data.to_csv('data_proses.csv', index=False)  # Menyimpan data yang telah diproses ke file CSV baru
```
